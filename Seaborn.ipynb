# Seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
datasets=sns.get_dataset_names()
print(datasets)
```

```python
data = sns.load_dataset("titanic")
print(data.head())
```

```python
sns.countplot(data,x='sex')
plt.title("How many male and feamle people are travelling in titanic")
```

```python
sns.displot(data,x='age',kde=True)
plt.title("Age distribution of people")

# kde= kernel distribution estimation it shows the data is uniform or normal distribution
```

```python
sns.barplot(data,x='sex',y='survived')
plt.title("Male and Female survived in titanic crash")
```

```python
sns.lineplot(data,x='class',y='fare')
plt.title("Class wise fare for titanic")
```

```python
sns.boxplot(data,x='age',y='alive')
plt.legend()
```

```python
sns.violinplot(data,y='survived')
```

```python
sns.stripplot(data,x='pclass',y='survived')
```

```python
sns.scatterplot(data,x='pclass',y='survived',hue='pclass')
```

```python
sns.pairplot(data,hue='sex')
```

```python
corr=data.select_dtypes(include='number').corr()
corr
```

```python
sns.heatmap(corr)

# relationship b/w numericals values
```
