```python
players=['Dhoni','Jadeja','Ruturaj','Dube']
runs=[51,50,100,87]
plt.bar(players,runs,color='green')
plt.title("Score Card")
plt.ylabel('runs')
plt.xlabel('Players')
plt.show()
```

```python
countries =['USA','India','China','Russia','Germany']
bronzes=np.array([38,17,26,19,15])
silvers=np.array([37,23,18,18,10])
golds=np.array([46,27,26,19,17])
#ind=[x for x, _ in enumerate(countries)]
plt.bar(countries,golds,label='golds',color='gold',bottom=silvers+bronzes)
plt.bar(countries,silvers,label='silvers',color='silver',bottom=bronzes)
plt.bar(countries,bronzes,label='bronzes',color='#CD853F')
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.legend()
plt.show()
```

```python
players='Dhoni','Jadeja','Ruturaj','Dube'
runs=[45,30,100,67]
plt.pie(runs,labels=players,autopct='%1.1f%%')
plt.show()
```

```python
players='Dhoni','Jadeja','Ruturaj','Dube'
runs=[45,30,100,67]
explode=(0,0,0.1,0)
# to make slices of pie chart with clockwise
plt.pie(runs,labels=players,autopct='%1.1f%%',explode=explode,
       shadow=True,startangle=90)
plt.show()
```

```python
# MARKERS
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o')
plt.show()

"""Markers
'o'- circle, '*'-star, '.'-point,','-pixel,'x'-X,'X'-X(filled),
'+'-plus,'s'-square,'D'-diamond
"""
```

```python
# LINE REFERNCES

# ms=marker size , the size of marker

# mec=marker color, marker color only shape outside

# mfc=marker face color , marker fill color

# linestyle = dotted,dashed,solid, dashdot

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=10,mec='r')
plt.grid()
plt.show()

x=np.array([3,8,1,10])
plt.plot(x,marker='D',ms=10,mfc='g',mec='g',
        linestyle='dotted')
plt.grid()
plt.show()
```

## multiple plots

```python
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y,marker='d',mec='g',mfc='g')
plt.title('sales')

x1=np.array([0,1,2,3])
y1=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x1,y1,marker='+',mec='r',mfc='r')
plt.title('profit')
plt.suptitle("SHOP")
plt.show()

# suptitle for all graphs one title
```

## Color map

```python
# Scatter plot

x=np.array([2,3,5,7,9,10,12,16,18,35,20,56,12,45,17])
y=np.array([5,7,8,9,10,15,67,89,35,27,30,46,28,12,27])

color = np.linspace(0, 100, len(x))

plt.scatter(x, y, c=color, cmap='viridis',s=44,alpha=0.5)
plt.colorbar()
plt.show()

# s=size of dot
# transparency = alpha=0.5
```

```python
data=data.drop_duplicates()
```

```python
plt.subplot(1,2,1)
plt.bar(x=data['Gender'],height=data['Income'],color='green')
plt.xlabel("Gender")
plt.ylabel("Income")
plt.show()

plt.subplot(1,2,2)
plt.bar(x=data['MaritalStatus'],height=data['Income'],color='blue')
plt.xlabel("Martial status")
plt.ylabel("Income")
plt.show()
```

```python
plt.hist(x=data['Age'],color='green',bins=5,histtype='stepfilled',alpha=0.8)
plt.xlabel("Age")
plt.title("Age distribution")
plt.show()
```

```python
plt.boxplot(data['Age'],patch_artist=True, boxprops=dict(facecolor='cyan', color='blue'))
plt.title("Age distribution")
plt.show()
```

```python
plt.violinplot(data['Age'])
plt.title("Age distribution")
plt.show()
```
